### 19.Analyse the text reviews count and show it on the count plot.

```python
bins=[0,10000,20000,30000,40000,50000,60000,70000,80000,90000,100000]
bin_name = ['0-10000','10000-20000','20000-30000','30000-40000','40000-50000','50000-60000','60000-70000','70000-80000','80000-90000','90000-100000']
df['text rating range']=pd.cut(df['text_reviews_count'],bins=bins,labels=bin_name)
df.head()
```

```python
df[''].value_counts()
```

```python
fig_dims =(30,12)
fig, ax= plt.subplots(figsize=fig_dims)
sns.countplot(y=df['text rating range'],ax=ax,data=df)
```
